```python
import json
import matplotlib.pyplot as plt

# JSON dosyasını yükle
with open('test_results.json', 'r') as file:
    test_results = json.load(file)

# Verileri ayrıştır
image_names = [result['image'] for result in test_results]
processing_times = [result['processing_time'] for result in test_results]

# İşleme sürelerini görselleştir
plt.figure(figsize=(10, 6))
plt.bar(image_names, processing_times, color='skyblue')
plt.title('Görsellerin İşleme Süresi')
plt.xlabel('Görsel Boyutu')
plt.ylabel('Süre (saniye)')
plt.xticks(rotation=45)
plt.show()
```

```python
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# JSON dosyasını yükle
with open('parametre_test_results.json', 'r') as file:
    test_results = json.load(file)

# Verileri Pandas DataFrame'e dönüştür
df = pd.DataFrame(test_results)

# İşleme sürelerini görselleştir
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='k_value', y='processing_time', hue='blur_kernel')
plt.title('Parametrelere Göre İşleme Süresi')
plt.xlabel('Küme Sayısı (k)')
plt.ylabel('Süre (saniye)')
plt.legend(title='Blur Kernel')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# CSV dosyasını yükleme
data = pd.read_csv('performance_results.csv')

# Görseller için işlem sürelerini görselleştirme
plt.figure(figsize=(12, 6))
for col in ['load_time', 'blur_time', 'lab_time', 'kmeans_time']:
    plt.plot(data['image'], data[col], label=col)

plt.plot(data['image'], data['total_time'], label='total_time', linestyle='--', linewidth=2, color='black')

plt.xlabel('Images')
plt.ylabel('Time (s)')
plt.title('Processing Times per Step')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV dosyasını yükleme
data = pd.read_csv('parameter_results.csv')

# k değerlerine göre filtreleme
k_values = data['k'].unique()

plt.figure(figsize=(12, 6))

for k in k_values:
    subset = data[data['k'] == k]
    plt.plot(subset['image'], subset['total_time'], label=f'k={k}')

plt.title('Processing Time for Different k Values')
plt.xlabel('Images')
plt.ylabel('Total Time (s)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV veya JSON'dan test sonuçlarını yükleme
test_results = pd.read_csv("test_results.csv")  # Eğer JSON ise: pd.read_json("test_results.json")

# Test sonuçlarını gruplandır ve özet çıkar
test_summary = test_results["result"].value_counts()

# Görselleştirme
plt.figure(figsize=(12, 6))
test_summary.plot(kind="bar")
plt.title("Test Sonuçlarının Özeti")
plt.xlabel("Test Durumu")
plt.ylabel("Test Sayısı")
plt.xticks(rotation=90)  # X ekseni etiketlerini döndür
plt.show()
```
