```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV dosyasını yükleme
data = pd.read_csv('parameter_results.csv')

# Veriyi inceleme
print(data.head())
```

```python
# k değerlerine göre filtreleme
k_values = data['k'].unique()

plt.figure(figsize=(12, 6))

for k in k_values:
    subset = data[data['k'] == k]
    plt.plot(subset['image'], subset['total_time'], label=f'k={k}')

plt.title('Processing Time for Different k Values')
plt.xlabel('Images')
plt.ylabel('Total Time (s)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Kernel boyutlarına göre filtreleme
kernel_sizes = data['kernel_size'].unique()

plt.figure(figsize=(12, 6))

for kernel in kernel_sizes:
    subset = data[data['kernel_size'] == kernel]
    plt.plot(subset['image'], subset['total_time'], label=f'Kernel={kernel}')

plt.title('Processing Time for Different Kernel Sizes')
plt.xlabel('Images')
plt.ylabel('Total Time (s)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
